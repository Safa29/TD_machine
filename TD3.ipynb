```python
n=2000 #inductor spires
d=0.5*10**-3 # wire diameter
e= 2*2.5*10**-3 #airgap 
u0=4*np.pi*10**-7
```

Now let's calculte the current in the inductor needed to creat a force of 100N. 

To generate 100N we will need an induction of:
\begin{equation}
 B=\sqrt{\frac{2\cdot F\cdot \mu_0}{S}}
 \end{equation}


```python
F=100
S=2*30*40*10**-6
B=np.sqrt(2*F*u0/S)
print (B)
```

In the airgap, this induction correspond to a magnetomotrice force of
 \begin{equation}
 AT_e=\frac{B \cdot e}{\mu_0}
 \end{equation}

```python
ATair=B*e/u0;
print (ATair)
```

In the steel, the magnetomotrice froce is given by the graph were we can read for 0.32T we will need 200AT/m. hence for our magnetique circuit we have: $ AT_{fer}=L\cdot 200\ AT$ where L is the length of the magnetic path.



TDIII.2.png

```python
#the length is provided in the upper figure. Remember the magnitic lines will take the middle path
L=(15+85+120)*2*10**-3

ATfer=L*200
print (ATfer)
```

The needed magnetomotrice force is

```python
ATtotal=ATair+ATfer
print (ATtotal)
```

The needed current is: $i=ATtotal/n$.  n: number of turn

```python
i=ATtotal/n
print (i)
```
